```python
df=df[df['Salary Estimate']!='-1']
```

```python
df['Salary Estimate']
```

```python
salary=df['Salary Estimate'].apply(lambda x:x.split('(')[0])
```

```python
salary
```

```python
salary=salary.apply(lambda x:x.replace('$','').replace('K',''))
```

```python
salary
```

```python
df['Min_Salary']=salary.apply(lambda x:x.split('-')[0])
```

```python
df['Max_Salary']=salary.apply(lambda x:x.split('-')[1])
```

```python
df
```

```python
df['Min_Salary'].unique()
```

```python
df['Min_Salary']=df['Min_Salary'].apply(lambda x:x.replace('Employer Provided Salary:',''))
```

```python
df['Min_Salary'].unique()
```

```python
df['Max_Salary'].unique()
```

```python
df['Max_Salary']=df['Max_Salary'].apply(lambda x:x.replace('Per Hour',''))
```

```python
df['Max_Salary'].unique()
```

```python
df['Min_Salary'].dtype
```

```python
df
```

```python
df['Min_Salary']=df['Min_Salary'].astype(int)
```

```python
df['Max_Salary']=df['Max_Salary'].astype(int)
```

```python
df['Average Salary']=(df['Min_Salary']+df['Max_Salary'])/2
```
